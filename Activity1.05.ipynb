<font color='orange'>  Plot Pearson Correlation Heatmap. </font>

```python
# 
data_num=data.drop(['state','gender','ic2'],axis=1) #Vamos a eliminar las columnas no numericos
data_num
```

```python
#Calcula la Correlacion
corr_matrixp=data_num.corr(method='pearson')
corr_matrixp
```

```python
plt.figure(figsize=(16, 6)) #Incrementar el TamaÃ±o del Heatmap
heatmap1=sns.heatmap(corr_matrixp,annot=True, cmap='cool') #annot=True para mostrar los valores de correlacion y cambiamos el color
heatmap1.set_title('Pearson Correlation Heatmap', fontdict={'fontsize':20}, pad=10,color='grey') #Le ponemos titulo 
```

```python
#definimos la mascara para establecer los valores del traingulo superior como true
mascara=np.triu(np.ones_like(data_num.corr(), dtype=np.bool)) # aislar el triÃ¡ngulo superior de una matriz mientras convierte todos los valores en el triÃ¡ngulo inferior en 
heatmap = sns.heatmap(data_num.corr(), mask=mascara, vmax=1, annot=True, cmap='cool') #podemos establecer el valor maximo y minimo
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':18}, pad=16, color='grey');
```

<font color='orange'> Plot Spearman Correlation Heatmap. </font>

```python
#Calcula la correlacion excluyendo los valores nulos
corr_matrixs=data_num.corr(method='spearman')
corr_matrixs
```

```python
plt.figure(figsize=(16, 6)) #Incrementar el TamaÃ±o del Heatmap
heatmap2=sns.heatmap(corr_matrixs,annot=True, cmap='twilight_shifted') #annot=True para mostrar los valores de correlacion y cambiamos el color
heatmap2.set_title('Spearman Correlation Heatmap', fontdict={'fontsize':20}, pad=10,color='grey') #Le ponemos titulo 
```
