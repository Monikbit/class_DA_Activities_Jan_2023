Try to convert these functions into lambda functions.


```python
#Funcion 1
def function_1(a,b):
    c=a+b
    return c
function_1(1,2) #Ejemplo
```

```python
#lambda
c=lambda a, b : a+b
c(1,2)    #Ejemplo
```

```python
#Funcion 2
def function_2(a,b):
    c=0
    for i in range(a):
        c+=2*b   # c += 2*b o c = c + 2*b
    return c
function_2(6,9)
```

```python
# Que pasa si c no empieza en cero?
c=lambda a,b : 2*b*a
c(6,9)
```

```python
#otra manera de hacerlo
c2=lambda a,b : sum([2*b for i in range (a)])
```

```python
c2(6,9)
```

```python
#Primero hay que pensar que me funcione para una entrada
def clean(state_name):
    state_name=state_name.upper()
    state_name=state_name.strip()
    if 'ARI' in state_name:
      return 'AR'
    elif 'TENN' in state_name:
      return 'TN'
    elif 'CALI' in state_name:
      return 'CA'
    else:
      return state_name
    
```

```python
var='Arizona'
clean(var)
```

```python
#Aplicar la funcion ya a toda la columna y pasar los outputs a una lista
#.map solo se puede usar en  variables y pbjetos de panda 
#map() es de python y te pide lista
#para poder guardar el resultado del map, casi siempre tiene que llevar el list
data2['state']=list(map(clean,data2['state']))
data2['state']
```
